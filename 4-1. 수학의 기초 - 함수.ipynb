# 1. 다항함수

$$
f(x) = 3x^2 + 2
$$
$$
$$

$$

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
x = np.linspace(-10, 10, 100000)
y = 3 * x**2 + 2
```

```python
x
```

```python
mpl.style.use('seaborn-whitegrid')

plt.figure(figsize=(12, 8))
plt.plot(x, y)
plt.xlabel('$x$', fontsize=25)
plt.ylabel('$y = 3x^2 + 2$', fontsize=18)
plt.show()
```

## 다항함수의 x축 방향 이동
$$
y_1  = 3x^2 + 2
$$

$$
y_2  = 3(x+1)^2 + 2
$$

- y2는 y1dmf x축 방향으로 -1 만큼 평행이동 한 것

```python
x = np.linspace(-5, 5, 1000)
y1 = 3 * x**2 + 2
y2 = 3 * (x+1)**2 + 2
```

```python
mpl.style.use('seaborn-whitegrid')

plt.figure(figsize=(12, 8))
plt.plot(x, y1, lw=2, ls='dashed', label='$y1 = 3x^2+2$')
plt.plot(x, y2, 'r', label='$y2 = 3(x+1)^2+2$')
plt.legend(fontsize=15)
plt.xlabel('$x$', fontsize=25)
plt.ylabel('$y$', fontsize=25)
plt.show()
```

# 2. 지수함수
$$
f(x) = a^x
$$
$$
$$
$$
a: 밑수
$$
$$
x : 지수
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2, 2, 100)
a11 = 2
a12 = 3
a13 = 4

y11 = a11**x
y12 = a12**x
y13 = a13**x
```

```python
a21 = 1/2
a22 = 1/3
a23 = 1/4

y21 = a21**x
y22 = a22**x
y23 = a23**x
```

```python
fig , ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].plot(x, y11, color='k', label='$2^x$')
ax[0].plot(x, y12,'--', color='k', label='$3^x$')
ax[0].plot(x, y13, ':', color='k', label='$4^x$')
ax[0].legend(fontsize=20)

ax[1].plot(x, y21, color='k', label='$(1/2)^x$')
ax[1].plot(x, y22,'--', color='k', label='$(1/3)^x$')
ax[1].plot(x, y23, ':', color='k', label='$(1/4)^x$')
ax[1].legend(fontsize=20)
```

- 밑수가 정수인 경우, 0 에서 시작해 한없이 커진다.
- 밑수가 정수인 경우, 밑수가 클수록 그래프가 크다.
- 밑수가 유리수인 경우, 0 으로 수렴한다.
- 밑수가 유리수인 경우, 밑수가 클수록 그래프가 작다.

```python
# 지수 증가

x = np.linspace(0, 10)

plt.figure(figsize=(6,6))
plt.plot(x, x**2, '--', color='k', label='$x^2$')
plt.plot(x, 2**x,  color='k', label='$2^x$')
plt.legend(loc="center left", fontsize=25)
plt.xlabel('$x$', fontsize=25)
plt.ylabel('$y$', fontsize=25)
plt.show()
```

## 자연상수 e
$
\displaystyle\lim_{x \to \infty}{(1+\frac{1}{x})^x} = \displaystyle\lim_{x \to \infty}{(1+\frac{1}{x})^{1/x}} = e
$
- 베르누이가 2.7182818 이란 숫자가 있다는걸 밝히고
- 오일러가 e 라는 표기를 시작함

# 3. 로그함수
$$
f(x) = log_a x
$$
$$
$$
$$
a : 밑수
$$

```python
# np.log는 밑수가 e라서 로그함수를 따로 만듦

def log(x, base):
    return np.log(x)/np.log(base)
```

```python
x1 = np.linspace(0.0001, 5, 1000)
x2 = np.linspace(0.01, 5, 1000)

y11 = log(x1, 10)
y12 = log(x2, np.e)
y21 = log(x1, 1/10)
y22 = log(x2, 1/np.e)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].plot(x1, y11, label='$\log_{10} x$', color='k')
ax[0].plot(x2, y12, '--', label='$\log_{e} x$', color='k')

ax[0].set_xlabel('$x$', fontsize=25)
ax[0].set_ylabel('$y$', fontsize=25)
ax[0].legend(fontsize=20, loc='lower right')

ax[1].plot(x1, y21, label='$\log_{1/10} x$', color='k')
ax[1].plot(x2, y22, '--', label='$\log_{1/e} x$', color='k')

ax[1].set_xlabel('$x$', fontsize=25)
ax[1].set_ylabel('$y$', fontsize=25)
ax[1].legend(fontsize=20, loc='upper right')

plt.show()
```

# 4. Sigmoid(시그모이드) 함수

$
\sigma(z) = \frac{1}{1+e^{-z}}
$

- 시그모이드는 0 과 1 사이의 값만 갖는다

```python
z = np.linspace(-10, 10, 100)
sigma = 1/(1 + np.exp(-z))

plt.figure(figsize=(12, 8))
plt.plot(z, sigma)
plt.xlabel('$z$', fontsize=25)
plt.ylabel('$\sigma(z)$', fontsize=25)
plt.show()
```

- 시그모이드는 0 과 1 사이의 값만 갖는다

```python
u = np.linspace(0, 1, 30)
v = np.linspace(0, 1, 30)

U, V = np.meshgrid(u, v)
```

```python
Z = (1 + U**2) + V / (1 + V**2)
```

```python
fig = plt.figure(figsize=(7, 7))
ax = plt.axes(projection='3d')  # 3차원 그래프를 그릴때 사용하는 속성

ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
ax.zaxis.set_tick_params(labelsize=15)

ax.set_xlabel('$u$', fontsize=20)
ax.set_ylabel('$v$', fontsize=20)
ax.set_zlabel('$z$', fontsize=20)

ax.scatter3D(U, V, Z, marker='.', color='gray')
plt.show()
```

# 7. 합성함수

```python
x = np.linspace(-4, 4, 100)
y = x**3 - 15*x + 30    # y = f(x)
z = np.log(y)    # z = g(y)
```

```python
# 두 그래프를 따로 그려보자

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].plot(x, y, label='$x^3 - 15x + 30$', color='k')
ax[0].legend(fontsize=18)

ax[1].plot(y, z, label='$\log(y)$', color='k')
ax[1].legend(fontsize=18)

plt.show()
```

```python
# 합성함수의 그래프를 그려보자

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].plot(x, z, '--', label='$log(f(x))$', color='k')
ax[0].legend(fontsize=18)

ax[1].plot(x, y, label='$\log(y)$', color='k')
ax[1].legend(fontsize=18)

ax_tmp = ax[1].twinx()
ax_tmp.plot(x, z, '--', label='$\log(f(x))$', color='k')

plt.show()
```
