### sklearn을 이용한 Decision Tree 구현
- fit(학습데이터, 정답) : 제공받은 데이터를 학습하는 모듈
- predict(데이터) : 학습한 내용을 바탕으로 데이터를 예측
- accuracy_score(정답, 예측값) : 두 데이터를 비교해 정확률 계산

````python
from sklearn.tree import DecisionTreeClassifier

iris_tree = DecisionTreeClassifier()
iris_tree.fit(iris.data[:, 2:], iris.target)
````

````python
# 성능 확인하기

from sklearn.metrics import accuracy_score

y_pred_tr = iris_tree.predict(iris.data[:, 2:])
y_pred_tr
````

````python
accuracy_score(iris.target, y_pred_tr)
````

````python
# Decision Tree를 시각화

from sklearn.tree import plot_tree

plt.figure(figsize=(12, 8))
plot_tree(iris_tree);
````

## 4. 과적합(overfitting)
- 지도학습으로 인해 과적합 발생
- 내가 가진 데이터에 너무 맞게 학습해서 다른 데이터에서는 성능이 낮게 나오는 것
- 그래서 accuracy가 높아도 좋게만 생각할수는 없다

- 데이터를 전부 모델학습에 사용하지 않고, 훈련(training), 검증(validation), 테스트(testing)으로 분리한다.

````python
# 처음부터 다시 시작!

from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
````

````python
# 데이터를 훈련/테스트 로 분리

from sklearn.model_selection import train_test_split

features = iris.data[:, 2:]
labels = iris.target

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=13) 
     # 8:2 비율로 feature(특성)과 labels(정답) 을 분리
````

````python
X_train.shape, X_test.shape
````

````python
# unique로 분리가 잘 되었는지 검사 

import numpy as np

np.unique(y_test, return_counts=True)
````

```
문제가 각 클래스(품종) 별로 동일 비율이 아닌 것.
큰 문제는 아니지만, 비율을 맞춰주는게 좋긴 하다
```

````python
# stratify 옵션으로 비율 맞춰주기

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, stratify=labels, random_state=13) 

np.unique(y_test, return_counts=True)
````

````python
# train 데이터로 결정나무 모델 만들기

from sklearn.tree import DecisionTreeClassifier
iris_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
iris_tree.fit(X_train, y_train)  # 학습
````

```
max_depth 를 설정하면 성능이 나빠진다.
depth를 설정하지 않으면, 모델이 나의 데이터에 근접하게 다가가는 문제가 발생한다.

학습할때 마다 일솬성을 위해 random_state를 고정...이해 못함

```

````python
# Decision Tree를 시각화

from sklearn.tree import plot_tree
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family = "Malgun Gothic")
%matplotlib inline

plt.figure(figsize=(10,8))
plot_tree(iris_tree);
````

````python
# 성능 확인하기

from sklearn.metrics import accuracy_score

y_pred_train = iris_tree.predict(X_train)
accuracy_score(y_train, y_pred_train)
````

````python
# mlxtend를 이용해 결정나무 모델이 어떻게 데이터를 분류했는지 확인

from mlxtend.plotting import plot_decision_regions

plt.figure(figsize=(14, 8))
plot_decision_regions(X=X_train, y=y_train, clf=iris_tree, legend=2)
plt.show();
````

````python
# 테스트 데이터에 대한 정확성

y_pred_test = iris_tree.predict(X_test)
accuracy_score(y_test, y_pred_test)
````

```
test 데이터에서 정확성이 더 높게 나왔다 -> train 데이터에 대한 과적합이 일어나지 않았다.
```

````python
# mlxtend를 이용해 결정나무 모델이 어떻게 데이터를 분류했는지 확인


plt.figure(figsize=(14, 8))
plot_decision_regions(X=X_test, y=y_test, clf=iris_tree, legend=2)
plt.show();
````

````python
scatter_highlight_kwargs= {'s':150, "label":"Test Data", "alpha":0.9}
scatter_kwargs = {'s':120, "edgecolor":None, "alpha":0.9}

plt.figure(figsize=(12, 8))
plot_decision_regions(X=features, y=labels, X_highlight=X_test, 
                      clf=iris_tree, legend=2, 
                      scatter_highlight_kwargs=scatter_highlight_kwargs, scatter_kwargs=scatter_kwargs,
                     contour_kwargs={"alpha":0.2})
````

````python
# feature를 4개 사용하는 경우


features = iris.data
labels = iris.target

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, stratify=labels, random_state=13)

iris_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
iris_tree.fit(X_train, y_train)
````

## 번외) zip 과 언패킹

### 1. 리스트를 튜플로 zip

````python
ls1 = ['a', 'b', 'c']
ls2 = [1, 2, 3]
````

````python
pairs = [pair for pair in zip(ls1, ls2)]
pairs
````

### 2. 다시 dic으로 변경

````python
dict(pairs)
````

### 3. 한번에 dict으로 만들기

````python
dict(zip(ls1, ls2))
````

### 4. unpacking인자를 이용한 역변환
- '*' 이 unpacking인자

````python
a, b = zip(*pairs)

print(a), print(b)   # 튜플로 반환된다.
````

````python
print(list(a))   # list() 를 사용해 다시 리스트형으로 변환
print(list(b))
````
